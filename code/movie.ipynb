```python
from scipy import stats
def init(N,Nc,sm):
    pci=np.zeros(shape=(N,Nc))
    for i in range(N):
        pci[i,0:Nc]=np.random.random_sample(Nc)
        pci[i,0:Nc]=pci[i,0:Nc]/sum(pci[i,0:Nc])
    
    Pc=np.zeros(Nc)
    Pc=pci.sum(axis=0)*(1/float(N))
    
    pic=np.zeros(shape=(N,Nc))
    pic=pci*(1/float(N))/Pc
    
    sci=np.zeros(shape=(N,Nc))
    for c in range(Nc):
        for m in range(N):
            for n in range(N):
                sci[m][c]+=pic[n][c]*float(sm[n][m])
    
    sc=np.zeros(Nc)
    resu_multi=np.multiply(pic,sci)
    sc=np.sum(resu_multi,axis=0)
    
    return pci,Pc,pic,sc,sci

def update_Pc(pci,Pc):
    Pc=pci.sum(axis=0)*(1/float(N))
    return Pc

def update_pic(pci,Pc,pic):
    pic=pci*(1/float(N))/Pc
    return pic

def update_sci(sm,pic,sci):
    for c in range(Nc):
        for i in range(N):
            sci[c][i]=0
            for k in range(N):
                sci[i][c]+=pic[k][c]*float(sm[k][i])
    return sci

def update_sc(sm,sc,pic):
    resu_multi=np.multiply(pic,sci)
    sc=np.sum(resu_multi,axis=0)
    return sc

def new_pci(pci,Pc,sci,sc,i):
    for c in range(Nc):
            sub_cal=float(div_T)*(2*sci[i][c]-sc[c])
            pci[i][c]=Pc[c]*math.exp(sub_cal)
    return pci

def update_pci(pci,i):

    sum_value=float(sum(pci[i,0:Nc]))
    for c in range(Nc):
        pci[i][c]=pci[i][c]/sum_value
    return pci

def juge_convege(pci,pci_last,epsilon):
    flag=True
    for c in range(Nc):
        for i in range(N):
            if (pci[i][c]-pci_last[i][c]>epsilon):
                flag=False
    return flag

import math
def algo(sm,N,Nc,epsilon,div_T):
    pci,Pc,pic,sc,sci=init(N,Nc,sm)
    pci_last=pci
    convege=False
    m=0
    while not convege:
        for i in range(N):
            pci=new_pci(pci,Pc,sci,sc,i)
            pci=update_pci(pci,i)
        flag=juge_convege(pci,pci_last,epsilon)
        m+=1
        if flag:
            convege=True
        else:
            Pc=update_Pc(pci,Pc)
            pic=update_pic(pci,Pc,pic)
            sci=update_sci(sm,pic,sci)
            sc=update_sc(sm,sc,pic)
            pci_last=pci
    counts=np.argmax(pci,axis=1)
    index=0
    result=[]
    for i in range(Nc):
        locals()['l'+str(i)]=[]
    for element in counts:
        for i in range(Nc):
            if (element==i):
                locals()['l'+str(i)].append(index)
                break
        index+=1
    for i in range(Nc):
        result.append(locals()['l'+str(i)])
    result=np.asarray(result)
    return result
```

```python
def get_matrix(filename):
    f=open(filename,"r")
    data=[]
    for line in f:
        data.append(line)
    f.close()
    matrix=[]
    for i in range(len(data)):
        clone=data[i].split(" ")
        for i in range(len(clone)):
            clone[i]=float(clone[i])
        matrix.append(clone)
    return matrix
```

```python
from scipy.spatial import distance
def calcute_distance(measure_similarity):
    d=np.zeros(Nc)
    for c in range(Nc):
        cluster=measure_similarity[c]
        d[c]=0
        for i in range(len(cluster)):
            for j in range(len(cluster)):
                d[c]+=distance.correlation(np.array(measure_similarity[c][i]).astype(float),np.array(measure_similarity[c][j]).astype(float))
        d[c]=d[c]/float(len(cluster))
    return d
```

```python
from operator import itemgetter
d=calcute_distance(measure_similarity)
real_dist=[]
for i in range(Nc):
    real_dist.append((i,d[i]))

real_dist=sorted(real_dist,key=itemgetter(1))
```
